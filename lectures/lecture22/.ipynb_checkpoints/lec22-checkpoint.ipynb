## Prediction

```python
galton = Table.read_table('galton.csv')
galton.show(10)
```

```python
galton.scatter('midparentHeight', 
              'childHeight')
```

```python
def predict_child(new_midparent_height):
    close_points = galton.where('midparentHeight', 
                                are.between(new_midparent_height - 0.5, 
                                            new_midparent_height + 0.5))
    return close_points.column('childHeight').mean()    
```

```python
# Apply predict_child to all the parent averages

height_pred = galton.with_column(
    'prediction', galton.apply(predict_child, 'midparentHeight')
)
height_pred
```

```python
height_pred.select('childHeight', 'midparentHeight', 'prediction')\
.scatter('midparentHeight')
```

```python
small = galton.sample(30)
small = small.with_column(
    'prediction', small.apply(predict_child, 'midparentHeight')
)
small.select('childHeight', 'midparentHeight', 'prediction').scatter('midparentHeight')
```

```python
small.select('childHeight', 'midparentHeight', 'prediction')\
.scatter('midparentHeight',fit_line=True)
```

## Quantify linearity

```python
# ignore this code
def r_scatter(r):
    "Generate a scatter plot with a correlation approximately r"
    plots.figure(figsize=(5,5))
    x = np.random.normal(0, 1, 1000)
    z = np.random.normal(0, 1, 1000)
    y = r*x + (np.sqrt(1-r**2))*z
    plots.scatter(x, y)
    plots.xlim(-4, 4)
    plots.ylim(-4, 4)
```

```python
r_scatter(1)
```

```python
r_scatter(0)
```

```python
r_scatter(-1)
```

```python
r_scatter(0.3)
```

```python
r_scatter(0.75)
```

## Computation of *r*
### Standard units

```python
def standard_units(numbers):
    "Convert any array of numbers to standard units."
    std = np.std(numbers)
    mean = np.mean(numbers)
    return (numbers - mean) / std
    #return (numbers - np.mean(numbers)) / np.std(numbers)  
```

```python
msrp = hybrid.column('msrp')
standard_units(msrp)
```

```python
# price vs. acceleration
hybrid.scatter('acceleration', 'msrp')

# price vs. acceleration in standard units
hybrid_msrp_su = hybrid.with_column(
    'msrp (su)', 
    standard_units(hybrid.column('msrp'))
)
hybrid_msrp_su.scatter('acceleration', 'msrp (su)')
```

**Q:** How do those two charts differ?

```python
# price vs. acceleration
hybrid.scatter('acceleration', 'msrp')

# price in standard units vs. acceleration in standard units
hybrid_msrp_accel_su = hybrid_msrp_su.with_column(
    'acceleration (su)', 
    standard_units(hybrid.column('acceleration'))
)
hybrid_msrp_accel_su.scatter('acceleration (su)', 'msrp (su)')
```

```python
def standardize(t):
    """Return a table in which all columns of t are converted to standard units."""
    t_su = Table()
    for label in t.labels:
        t_su = t_su.with_column(
            label + ' (su)', 
            standard_units(t.column(label))
    )
    return t_su
```

```python
accel_msrp_su = standardize(hybrid.select('acceleration', 'msrp'))
accel_msrp_su
```

```python
accel_msrp_su.scatter('acceleration (su)', 'msrp (su)')
```

```python
standardize(hybrid.select('mpg', 'msrp')).scatter('mpg (su)', 'msrp (su)', )
```

### A function to compute *r*

```python
def correlation(t, x, y):
    """The correlation coefficient of columns x and y in table t."""
    return np.mean(standard_units(t.column(x)) * standard_units(t.column(y)))
```

```python
correlation(hybrid.select('acceleration', 'msrp'), 'acceleration', 'msrp')
```

## Properties of correlation

**1. Unaffected by changing units**

```python
accel_msrp = hybrid.select('acceleration', 'msrp')
correlation(accel_msrp, 'acceleration', 'msrp')
```

```python
accel_msrp_su = standardize(accel_msrp)
correlation(accel_msrp_su, 'acceleration (su)', 'msrp (su)')
```

```python
accel_msrp_1k = accel_msrp.with_column(
    'msrp ($k)', accel_msrp.column('msrp') / 1000
)
correlation(accel_msrp_1k, 'acceleration', 'msrp ($k)')
```

Because we're measuring the **relative** spread around a straight line.

```python
accel_msrp.scatter('acceleration', 'msrp')
accel_msrp_su.scatter('acceleration (su)', 'msrp (su)')
accel_msrp_1k.scatter('acceleration', 'msrp ($k)')
```

**2. Unaffected by swapping variables**

```python
correlation(hybrid, 'acceleration', 'msrp')
```

```python
correlation(hybrid, 'msrp', 'acceleration')
```

Because the roles of $x$ and $y$ are symmetric in formula, and graphs are symmetric when you swap axes.

## Prediction using correlation

```python
galton = Table.read_table('galton.csv')
heights = galton.select('midparentHeight', 'childHeight')\
  .relabeled('midparentHeight', 'MidParent')\
  .relabeled('childHeight', 'Child')
heights.show(3)
```

```python
heights.scatter('MidParent')
```

```python
heights_su = standardize(heights)
heights_su.scatter('MidParent (su)')
```

```python
def predict_child_su(new_midparent_height):
    close_points = heights_su.where('MidParent (su)', 
                                are.between(new_midparent_height - 0.5, 
                                            new_midparent_height + 0.5))
    return close_points.column('Child (su)').mean()    

heights_su_pred = heights_su.with_column(
    'Prediction (su)', heights_su.apply(predict_child_su, 'MidParent (su)')
)
heights_su_pred.scatter('MidParent (su)')
```

The yellow line is the *graph of averages*.

**Q:** What is the predicted height (in su) of a child whose parents have an average height?

A. -1  
B. 0  
C. 1  
D. Whatever $r$ is for this scatter plot  
E. Can't answer from this plot  

<br/><br/><br/><br/><br/><br/>

```python
r = correlation(heights, 'MidParent', 'Child')
r
```

## The graph of averages and $r$

```python
heights_su_pred.scatter('MidParent (su)')

# plot the line "y = r * x"
plots.plot([-3, 3.5], [-3*r, 3.5*r], color='blue', lw=2);
```

The line $y = r \times x$ is a version of the graph of averages, smoothed to a line.

## Regression to the mean

```python
heights_su.scatter('MidParent (su)')

# Plot the "SD line"
plots.plot([-3, 3.5], [-3, 3.5], color='green', lw=2);
```

```python
heights_su.scatter('MidParent (su)')

# Plot the "SD line"
plots.plot([-3, 3.5], [-3, 3.5], color='green', lw=2);

# plot the regression line "y = r * x"
plots.plot([-3, 3.5], [-3*r, 3.5*r], color='blue', lw=2);
```

Regression line is flatter than SD line:  extreme values of $x$ result in not-as-extreme values of $y$
